# Q1. What is parameter?
-> A parameter is a variable in a model that is learned from the training data. In machine learning, parameters are interval values such as weights in a neural network or cofficients in a linear regression model.

# Q2. What is correlation. What does negative correlation mean?
-> Correlation is a statistical measure that describes the strength and direction of a relationship between two variables. It ranges from -1 to +1.
A negative correlation means that as one variable increases, the other decreases. For example, if the temprature increases , sales of winter jackets decreases, showing a negative correlation.

# Q3. Define Machine Learing. What are the main components in Machine Learning?
-> Machine Learning is a subset of artificial intelligence that enables computers to learn from data and make predictions or decisions without being explicitly programmed.
The main components of Machine Learning are :
- Data: The dataset used for training and testing.
- Features: The input variables used to train a model.
- Model: The mathematical representation that learns from data.
- Loss Function: A measure of how well the model is performing.
- Optimization Algorithm : Used to minimize the loss function, such as gradient descent.

# Q4. How does loss value help in determining whether the model is good or not?
-> The loss value quantifies how far the model's predictions are from the actual values. A lower loss value indicates a better model , while a higher loss suggests poor performance.

# Q5. What are continuous and categorial variables?
->
- $Continuous$ $variables$ are numerical values that can take an infinite number of values within a range (e.g., height, temprature).

- $Categorial$ $variables$ represent discrete groups or categories (e.g., gender ,color, product type).

# Q6. How do we handle categorical variables in Machine Learning ? What are the common techniques ?
-> Categorical variables can be handled using:
- One - Hot Encoding : Converts categories into binary columns.

- Label Encoding : Assigns numerical values to categories.

- Ordinal Encoding: Assigns ordered numerical values when categories have a hierarchy.

#Q15. What is causation? Explain the difference between correlation and causation with an example.
->
- $Causation$ means  that a change in one variable directly causes a change in another. Example: Ice cream sales and drowning incidents increases in summer.
- $Correlation$ means two variables move together,but one does not necessarily cause the other.
Example: High tempratures cause people to swim more, increasing drowning incidents.

#Q16. What is an Optimizer ? What are different types of optimizers? Explain each with an example.
-> An optimizer in machine learning adjusts the model's parameters to minimize the loss function.
Common optimizers:
- $Gradient Descent$ : Updates parameters in the direction of decreasing loss.
- $ Stochastic Gradient Descent (SGD)$: Updates parameters using a small batch of data at a time.
- $Adam (Adaptive Moment Estimation)$: Uses adaptive learning rates for better convergence.
